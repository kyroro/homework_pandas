## 2 задача

### 2.1 
Изучите распределение домов от наличия вида на набережную
Постройте график
Сделайте выводы

```python
data = df['waterfront'].value_counts()
names = data.index
values = data.values
plt.pie(values, autopct='%.1f%%', labels=names);
```

Только 0.8% домов имеют вид на набережную

### 2.2 
Изучите распределение этажей домов

```python
data = df['floors'].value_counts()
names = data.index
values = data.values
plt.figure(figsize=(8, 6))
plt.bar(names, values)
plt.xticks(rotation=50);
```

Большинство домов с одним этажом

### 2.2 
Изучите распределение состояния домов

```python
data = df['condition'].value_counts()
names = data.index
values = data.values
plt.figure(figsize=(8, 6))
plt.bar(names, values)
plt.xticks(rotation=50);
```

Большинство домов с оценкой качества 3, а меньше всего  1


## 3 задача
Исследуйте, какие характеристики недвижимости влияют на стоимость недвижимости, с применением не менее 5 диаграмм из урока.
Анализ сделайте в формате storytelling: дополнить каждый график письменными выводами и наблюдениями.

```python
corr_matrix = df.corr()
corr_matrix = np.round(corr_matrix, 1)
corr_matrix[np.abs(corr_matrix) < 0.3] = 0
corr_matrix
```

```python
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=True, linewidths=.5, cmap='coolwarm');
```

```python
fig, ax = plt.subplots(figsize=(10, 6))
ax.scatter(x = df['price'], y = df['sqft_living'])
plt.xlabel("price")
plt.ylabel("sqft_living")

plt.show()
```

Связь между ценой и $м^2$ жилья, коэффициент 0.7 - заметная

Так же наблюдается связь между ценой и количеством комнат, коэффициент 0.5 - слабая
